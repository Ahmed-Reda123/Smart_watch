```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('dailyActivity_merged.csv')
df
```

```python
df['ActivityDate'] = pd.to_datetime(df['ActivityDate'], format="%m/%d/%Y")
df.info()
```

```python
df['TotalMiutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes'] + df['LightlyActiveMinutes'] + df['SedentaryMinutes']
df.head(5)
```

```python
df.describe()
```

```python
sns.scatterplot(x='Calories', y='TotalSteps', data=df)
```

```python
labels = ['Very Active', 'Fairly Active', 'Lightly Active', 'Sedentary']
values = df[['VeryActiveMinutes', 'FairlyActiveMinutes', 'LightlyActiveMinutes', 'SedentaryMinutes']].mean()
colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=30,
                  marker=dict(colors=colors, line=dict(color='black', width=3)))
fig.update_layout(title_text='Daily Minutes', title_x=0.5)
fig.show()
```

```python
df['Day'] = df['ActivityDate'].dt.day_name()
df.head(5)
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["VeryActiveMinutes"],
    name='Very Active',
    marker_color='red'
))
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["FairlyActiveMinutes"],
    name='Fairly Active',
    marker_color='green'
))
fig.add_trace(go.Bar(
    x=df["Day"],
    y=df["LightlyActiveMinutes"],
    name='Lightly Active',
    marker_color='blue'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```
